```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1,1)
```

```python
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression 
linear_reg = LinearRegression()
linear_reg.fit(X, y)
predict_y = linear_reg.predict(X)
plt.scatter(x, y)
plt.plot(x, predict_y, color='r')
plt.show()
```

```python
(X**2).shape
```

```python
# 增加了一个纬度 X的平方
X2 = np.hstack([X, X**2])
```

```python
linear_reg2 = LinearRegression()
linear_reg2.fit(X2, y)
```

```python
# X2的确多了一个纬度，但是预测的结果还是和原来的y一样的，用两个纬度的X2预测y
y_predict2 = linear_reg2.predict(X2)
```

```python
plt.scatter(x, y)
# 用预测出来的y_predict2和原始的x画一条线
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r')
plt.show()
```

```python
# 系数，第一个是x的系数，第二个是x平方的系数
linear_reg2.coef_
```

```python
# 截距，与y轴的交点
linear_reg2.intercept_
```
