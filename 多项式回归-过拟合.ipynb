```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1,1)
```

```python
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression 
linear_reg = LinearRegression()
linear_reg.fit(X, y)
predict_y = linear_reg.predict(X)
plt.scatter(x, y)
plt.plot(x, predict_y, color='r')
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
# 多项式回归Pipeline
def PolynomialRegression(degree):
    return Pipeline([
                ('std_scaler', StandardScaler()),
        ('poly', PolynomialFeatures(degree=degree)),

        ('lin_reg', LinearRegression())
    ])
```

```python
poly_reg = PolynomialRegression(2)
poly_reg.fit(X, y)
y2_predict_std_first = poly_reg.predict(X)
mean_squared_error(y, y2_predict_std_first)
```

```python
poly_reg = PolynomialRegression(2)
poly_reg.fit(X, y)
y2_predict = poly_reg.predict(X)
```

```python
plt.scatter(x, y)
# 用预测出来的y_predict2和原始的x画一条线
plt.plot(np.sort(x), y2_predict[np.argsort(x)], color='r')
plt.show()
```

```python
# 二次方程的均方误差
mean_squared_error(y, y2_predict)
```

```python
# 线性回归的均方误差
mean_squared_error(y, predict_y)
```

```python
poly_reg3 = PolynomialRegression(30)
poly_reg3.fit(X, y)
y2_predict3 = poly_reg3.predict(X)
```

```python
# 三次方程的均方误差
mean_squared_error(y, y2_predict3)
```

```python
plt.scatter(x, y)
# 用预测出来的y_predict2和原始的x画一条线
plt.plot(np.sort(x), y2_predict3[np.argsort(x)], color='r')
plt.show()
```
